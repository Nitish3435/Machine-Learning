```python
fd=cv2.CascadeClassifier("CV2.xml")
```

```python
faces=fd.detectMultiScale(img,1.1,5)
```

```python
print(faces)
```

```python
face_0=faces[0]
x,y,w,h=face_0
print(face_0)
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x,y=make_blobs(n_samples=300, n_features=2,  centers=2, )
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**.5
def KNN(X,Y,test_point,k=5):
    m = X.shape[0]
    all_dist = []
    
    for i in range(m):
        d = dist(test_point,X[i])
        all_dist.append((d,Y[i]))
    all_dist.sort()
    a=np.array(all_dist[:k])
    print(a)
```

```python
KNN(x,y,5,5)
```
